```python
# Impute missing values with the mean for Glucose, BloodPressure, and BMI
columns_with_zeros = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
for column in columns_with_zeros:
    data[column] = data[column].replace(0, pd.NA)

# Ensure columns are numeric to avoid warnings
data[columns_with_zeros] = data[columns_with_zeros].astype(float)

# Impute missing values with the mean for Glucose, BloodPressure, and BMI
for column in ["Glucose", "BloodPressure", "BMI"]:
    data[column] = data[column].fillna(data[column].mean())

# Impute missing values with the median for SkinThickness and Insulin
for column in ["SkinThickness", "Insulin"]:
    data[column] = data[column].fillna(data[column].median())

# Verify there are no missing values left
print("Missing values after imputation:")
print(data.isna().sum())
print(data)
```

```python
# Min-Max Normalization
import seaborn as sns
df = data.drop('Pregnancies', axis=1)
df_norm = (df-df.min())/(df.max()-df.min())
df_norm = pd.concat((df_norm, data.Pregnancies),axis= 1)

print("Scaled Dataset Using Pandas")
df_norm.head()
```
